```python
dados = {"Ma": 1.50, 
"Banana": 2.75,
"Uva": 1.90, 
"Pera": 1.25, 
"Laranja": 0.65, 
"Limo": 1.25, 
"Goiaba": 2.15, 
"Abacaxi": 3.20, 
"Jaca": 5.80}

print(dados)
```

```python
dados.values()
```

```python
fruta = input("Digite o nome da fruta:")
print(fruta)
```

```python
if fruta in dados:
    print(f"A fruta {fruta} custa R$ {dados[fruta]:.2f}") # Formatao para float (duas casas decimais) = .2f
else:
    print(f"A fruta '{fruta}' no foi encontrada.")
```
