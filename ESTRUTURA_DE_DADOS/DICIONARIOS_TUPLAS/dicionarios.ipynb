Dicionrios em Python (`dict`)

* Um **dicionrio** armazena **pares chave\:valor**.
* As **chaves so nicas** e os valores podem ser de qualquer tipo.

Sintaxe:

```python
meu_dict = {
    "nome": "Ana",
    "idade": 25,
    "cidade": "So Paulo"
}
```



Operaes comuns:

**Acessar valor por chave**

```python
print(meu_dict["nome"])  # Sada: Ana
```

**Adicionar ou atualizar item**

```python
meu_dict["profisso"] = "Engenheira"
```

**Remover item por chave**

```python
del meu_dict["idade"]
```

**Verificar se chave existe**

```python
if "cidade" in meu_dict:
    print("Chave encontrada")
```

**Percorrer dicionrio**

```python
for chave, valor in meu_dict.items():
    print(chave, "->", valor)
```

**Obter apenas chaves ou valores**

```python
meu_dict.keys()     # retorna as chaves
meu_dict.values()   # retorna os valores
```



Exemplo:

```python
aluno = {"nome": "Lucas", "nota": 8.5}

print(f"{aluno['nome']} tirou nota {aluno['nota']}")
```

> Sada:

```
Lucas tirou nota 8.5
```




````python
# pares de chave/valor, exemplo:
# pode usar strings, inteiros, floats, etc. como chaves
# valores pode ser de qualquer tipo de dados

dados = {
    "nome": "Maria",
    "idade": 30,
    "cidade": "So Paulo",
    "profisso": "Engenheira",
    "hobbies": ["ler", "viajar", "cozinhar"],
    "ativo": True,
    "altura": 1.65,
    "peso": 60.5
    }

print(dados)
````

````python
dados["nome"]  # Acessando o valor pela chave
````

````python
dados.get("idade")  # Acessando o valor com get()
````

````python
dados.keys()    # retorna as chaves
````

````python
dados.values()  # retorna os valores
````

````python
for i in dados:
    print(i, ":", dados[i])  # Imprime chave e valor
````
