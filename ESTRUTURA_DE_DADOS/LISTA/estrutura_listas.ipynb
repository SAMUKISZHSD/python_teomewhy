Listas em Python

* Uma **lista** é uma coleção de itens **mutável** e **ordenada**.
* Pode conter **qualquer tipo de dado**: números, strings, outras listas, etc.

Sintaxe:

```python
minha_lista = [1, 2, 3, 4]
```

---

Operações básicas:

**Acessar itens (por índice)**

```python
print(minha_lista[0])  # Saída: 1
```

**Modificar um item**

```python
minha_lista[1] = 99
```

**Adicionar item ao final**

```python
minha_lista.append(5)
```

**Inserir em posição específica**

```python
minha_lista.insert(2, 100)  # posição 2, valor 100
```

**Remover item por valor**

```python
minha_lista.remove(99)
```

**Remover item por índice**

```python
del minha_lista[0]
```

**Tamanho da lista**

```python
len(minha_lista)
```

**Percorrer com for**

```python
for item in minha_lista:
    print(item)
```

---

Exemplo completo:

```python
frutas = ["maçã", "banana", "uva"]
frutas.append("laranja")
for fruta in frutas:
    print(fruta)
```

> Saída:

```
maçã  
banana  
uva  
laranja
```
