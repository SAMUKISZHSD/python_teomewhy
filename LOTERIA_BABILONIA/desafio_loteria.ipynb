```python
import random
```

```python
num_sorteiado = random.randint(1, 15)
print(f"Número sorteado: {num_sorteiado}")
```

```python
tentativas = 3
```

```python
print("Adivinhe o número entre 1 e 15. Você tem 3 tentativas.")
```

```python
for tentativa in range(1, tentativas + 1):
    chute = int(input(f"Tentativa {tentativa}: Digite seu palpite: "))
    
    if chute == num_sorteiado:
        print("Parabéns! Você acertou o número!")
        break
    elif chute < num_sorteiado:
        print("O número é maior que isso.")
    else:
        print("O número é menor que isso.")

else:
    print(f"Você perdeu! O número era {num_sorteiado}.")
```
