```python
import random
```

```python
num_sorteiado = random.randint(1, 15)
print(f"Nmero sorteado: {num_sorteiado}")
```

```python
tentativas = 3
```

```python
print("Adivinhe o nmero entre 1 e 15. Voc tem 3 tentativas.")
```

```python
for tentativa in range(1, tentativas + 1):
    chute = int(input(f"Tentativa {tentativa}: Digite seu palpite: "))
    
    if chute == num_sorteiado:
        print("Parabns! Voc acertou o nmero!")
        break
    elif chute < num_sorteiado:
        print("O nmero  maior que isso.")
    else:
        print("O nmero  menor que isso.")

else:
    print(f"Voc perdeu! O nmero era {num_sorteiado}.")
```
